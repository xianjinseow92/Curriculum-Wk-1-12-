# We're going to need bigger data!

In order to see important differences between models, we'll want to look at them as they handle datasets that are much bigger than what we've used so far in class.

In order to do that, we'll use the full MNIST dataset ([more info](http://yann.lecun.com/exdb/mnist/)). This classic dataset contains 70,000 images of hand written digits. Each image is 28x28 pixels. Just like we saw in the KNN digits demo, we'll perform image recognition by flattening each image so that each of the 784 pixels are treated like a feature.

In total, this gives us about **60MB of data**. It may not sound like much, but you'll see this amount **cause trouble** for **several of the algorithms** we've discussed.

```python
# MNIST is a very famous dataset on hand written digits

# To the computer, we are not dealing with the  color of the image.
# We are dealing with the COLOR CODE of the image. The RGB values, basically!
```

```python
# Load in the data and measure its size 

(x, y), (x_test, y_test) = keras.datasets.mnist.load_data() # We are just using keras to get the dataset

mb = (x.size + x_test.size + y.size + y_test.size)/1e6

print(f"dataset size: {mb}MB")
```

```python
# Reshape each observation to be 784 features instead of a 28x28 matrix 
x = preprocessing.MinMaxScaler().fit_transform(x.reshape(-1,784).astype("float64"))  # We are using a min/max scaler

# Reshape. We don't care about the number of rows. Let the function figure out how to fit your rows in
# -1 is basically letting the function decide how many rows to make


# Just remember that for images we HAVE to scale them



# MNIST isn't shuffled by default so let's shuffle x and y
indices = np.random.choice(len(y), size=len(y), replace=False)  # will return you the shuffled index
x = x[indices, :]; y = y[indices]
```

```python
# Let's import a few more algorithms so we can time time on different sized samples

from sklearn import svm, ensemble, linear_model, neighbors, naive_bayes # importing all the models
```

```python
# Has it been 30 seconds since we ran the last cell? This is how we'll keep track of how much time has passed
lap_time.shift(seconds=30) < arrow.now() 

# this will give us the time that we kept in arrow.now() + 30 seconds
```

```python
import timeit # We'll use this package to run each simulation multiple times and take the fastest runtime 

num_loops = 5
num_repeats = 1
num_tests = 20
max_samples = 1500  # observations
max_features = 100 # "pixels"
max_sec = 30

# my favorite way to build a datafrom on the fly is 
# to start with a list of dictionaries
complexity_lod = []


# Each time we run our algorithm, we'll want to use a larger sample size to test the effect of sample size on runtime
# Set up the list of sample sizes we'll use
samples = np.linspace(50,max_samples, num=num_tests, dtype=int)  # get number of samples from 50 to 1500


# We'll want to test multiple algorithms 
model_dict = {
    "svm-linearSVC": svm.LinearSVC(),  # SVC (Support Vector Classifier) is for classification (1 and 0's), SVM is for regression
    "svm-rbf": svm.SVC(gamma="auto"), # Reynold Basis Function (For when data is in circles/not linearly separable)
    "svm-poly3": svm.SVC(kernel="poly", gamma="auto"),  # Transforming your data
    "random-forest": ensemble.RandomForestClassifier(n_estimators=100),  
    "logistic": linear_model.LogisticRegression(solver="lbfgs", multi_class="auto"), # logistic regression
    "knn": neighbors.KNeighborsClassifier(),  # KNN
    "nb": naive_bayes.GaussianNB()  # GaussianNaiveBase because it is still some kind of continous variable even after scaling
}  # When we call a specific key in this dictionary, we are actually calling the model object


# Run all of our models, increasing the sample size each time (from samples list we made above)
for n_samp in samples:   # Store the number of samples 
    print(f"Running for {n_samp} samples. ", end="") 
    # for eg, if n_samples = 50 samples, will loop through each algorithm and train it on 50 samples
    lap_time = arrow.now()
    
    # Loop through each algorithm 
    for model_key in model_dict:
        
        model = model_dict[model_key]  # Initializing the model by calling the model key and accessing the value (which is the model object itself)
    
        # we take the min of the repeats because variation is likely due not to python execution 
        # but other operations https://docs.python.org/3/library/timeit.html#timeit.Timer.repeat
        train_time = np.min(
            timeit.repeat(
                f'model.fit(x[:n_samp,:max_features],y[:n_samp])',
                globals=globals(),
                repeat=num_repeats,
                number=num_loops))

        test_time = np.min(
            timeit.repeat(
                f'model.predict(x[:n_samp,:max_features])',
                globals=globals(),
                repeat=num_repeats,
                number=num_loops))

        # Storing the results from all the trainings into a dictionary, of which is then stored in a list
        complexity_lod.append({
                "model": model_key,
                "train_time": train_time,
                "test_time": test_time,
                "n_samples": n_samp
            })
        
    print(f"Time: {(arrow.now() -lap_time).seconds} seconds.")
    if lap_time.shift(seconds=max_sec) < arrow.now():  # We are giving it 30s to train. if it is taking more than 30s, we will break the training
        print("breaking")
        break
        
complexity_df = pd.DataFrame(complexity_lod)
```

```python
# This graph plots the time taken to train vs the number of observations to train

# This is for TRAINING
sns.lineplot(x="n_samples", y="train_time", hue="model", data=complexity_df)
```

```python
# We can see that in TESTING, our KNN is VERY VERY expensive when it comes to prediction.
# We will most likely use logistic, random-forest and NB
sns.lineplot(x="n_samples", y="test_time", hue="model", data=complexity_df)
```

# Takeaways

What does this tell us? These analyses don't give us hard-and-fast answers. We decided to teach these models at Metis because **every one of them is either important to understand or likely to be used in industry, and often both.** 

Two of these in particular are worth noting: **KNN and Random Forests**. Both of these models are strong performers on benchmark tasks, where we have a fixed dataset and we want to see which model performs best (these are often derisively referred to as "bake offs").

But (as they are implemented in sklearn) they both have a weakness: **Prediction time scales linearly with the number of examples in our training set**.  
  
This is **important** because often __*the only or easiest way to improve a model is to add more and better training data*.__  
  
For these two, **more data might lead to better predictions**, but also leads to **slower performance**. You'll find that this is always true of KNN but you might be able to tweak the hyperparameters of Random Forests to get improved performance.
